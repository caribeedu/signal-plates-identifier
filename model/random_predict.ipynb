## Random image prediction

```python
from mrcnn.visualize import display_instances
from mrcnn.config import Config

import numpy as np

from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array

from helpers import build_dataset, create_model

class PredictionConfig(Config):
    NAME = "predict"
    NUM_CLASSES = 1 + 51
    DETECTION_MIN_CONFIDENCE = 0.7
    GPU_COUNT = 1
    IMAGES_PER_GPU = 1

SETS_PATH = "./dataset"

val_set = build_dataset("val", SETS_PATH)

MODEL_WEIGHTS_PATH = './signal_plates_mask_rcnn.h5'

config = PredictionConfig()

model = create_model('inference', config, MODEL_WEIGHTS_PATH)
```

```python
image_ids = np.random.choice(val_set.image_ids, 2)

for image_id in image_ids:
  img = load_img(val_set.image_info[image_id]['path'])
  img = img_to_array(img)

  result = model.detect([img], verbose=0)[0]

  if len(result['class_ids']) == 0:
    continue

  display_instances(img, result['rois'], result['masks'], result['class_ids'], val_set.class_names, result['scores'])
```
